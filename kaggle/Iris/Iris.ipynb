Outlines different visual exploration of data using seaborn and pandas data visualisation 


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = pd.read_csv("Iris.csv")
```

```python
iris.head()
```

```python
iris["Species"].value_counts()
```

```python
iris.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm")
```

```python
# Show bivariate scatterplots and univariate histograms in the same chart
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=iris, size=5)
```

```python
# Use seaborn's FacetGrid to color the scatterplot by species
sns.FacetGrid(iris, hue="Species", size=5) \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend()
```

```python
# Explore an individual feature through a boxplot
sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
```

```python
# Explore an individual feature through a kdeplot,
# which creates and visualizes a kernel density estimate of the underlying feature
sns.FacetGrid(iris, hue="Species", size=6) \
   .map(sns.kdeplot, "PetalLengthCm") \
   .add_legend()
```

```python
# Another useful seaborn plot is the pairplot, which shows the bivariate relation
# between each pair of features
# 
# From the pairplot, we'll see that the Iris-setosa species is separataed from the other
# two across all feature combinations
sns.pairplot(iris.drop("Id", axis=1), hue="Species", size=3)
```

```python
# Create boxplot with Pandas on each feature split out by species
iris.drop("Id", axis=1).boxplot(by="Species", figsize=(12, 6))
```

```python
from pandas.plotting import radviz
radviz(iris.drop("Id", axis=1), "Species")
```
